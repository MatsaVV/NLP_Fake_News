```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt


# Séparer les vraies nouvelles ("True") et les fausses nouvelles ("Fake")
vraies_nouvelles = df[df['label'] == 'True']['text'].str.cat(sep=' ')
fausses_nouvelles = df[df['label'] == 'Fake']['text'].str.cat(sep=' ')

# Créer des nuages de mots
wordcloud_vraies = WordCloud(width=800, height=400, background_color='white').generate(vraies_nouvelles)
wordcloud_fausses = WordCloud(width=800, height=400, background_color='white').generate(fausses_nouvelles)

# Afficher le nuage de mots pour les vraies nouvelles
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_vraies, interpolation='bilinear')
plt.axis('off')
plt.title('Nuage de mots - Vraies Nouvelles')
plt.show()

# Afficher le nuage de mots pour les fausses nouvelles
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_fausses, interpolation='bilinear')
plt.axis('off')
plt.title('Nuage de mots - Fausses Nouvelles')
plt.show()
```

```python
import pandas as pd
import string

# Charger ton DataFrame (si ce n'est pas déjà fait)
# fake = pd.read_csv('chemin/vers/ton/fichier.csv')

# Fonction pour compter les occurrences de ponctuation
def count_punctuation(text, punctuation_chars):
    return sum(text.count(char) for char in punctuation_chars)

# Définir les caractères de ponctuation à compter
punctuation_chars = string.punctuation  # Cela inclut tous les signes de ponctuation

# Appliquer la fonction à chaque texte
for char in punctuation_chars:
    fake[f'{char}_count'] = fake['text'].apply(lambda x: x.count(char))

# Afficher les statistiques pour chaque caractère de ponctuation
for char in punctuation_chars:
    column_name = f'{char}_count'
    print(f"Nombre maximum de '{char}': {fake[column_name].max()}")
    print(f"Nombre moyen de '{char}': {fake[column_name].mean()}")
    print(f"Nombre minimum de '{char}': {fake[column_name].min()}")
```
